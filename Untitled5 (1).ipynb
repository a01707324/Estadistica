¿Qué probabilidad hay de que un cliente abandone basándonos en su nivel de satisfacción?


```r
library(ggplot2)
library(dplyr)
library(stats)

#Datos
df <- read.csv("Customer-Churn-Records.csv")

# Vectores Resultados
probChurnSatis <- numeric(5)
propSatis <- numeric(5)
probChurnSatisF <- numeric(5)
probChurnSatisM <- numeric(5)

for (i in 1:5) {
  satis <- df[df$Satisfaction.Score == i, ]
  probChurnSatis[i] <- sum(satis $Exited == 1) / nrow(satis)
  propSatis[i] <- nrow(satis) / nrow(df)
}
promedioProbSatis <- mean(df$Satisfaction.Score, na.rm = TRUE)
cat(paste0("Promedio de Satisfaction: ", round(promedioProbSatis, 4), "\n"))
cat ("\n")
promedioProbNoChurnSatis <- mean(df[df$Exited == 0, "Satisfaction.Score"], na.rm = TRUE)
cat(paste0("Promedio de Satisfaction Score No Churn: ", round(promedioProbNoChurnSatis, 4), "\n"))
cat ("\n")
promedioProbChurnSatis <- mean(df[df$Exited == 1, "Satisfaction.Score"], na.rm = TRUE)
cat(paste0("Promedio de Satisfaction Score Churn: ", round(promedioProbChurnSatis, 4), "\n"))
cat ("\n")

# Histograma
ggplot(df, aes(x = Satisfaction.Score)) +
  geom_histogram(binwidth = 1, fill = "steelblue", color = "black", boundary = 0.5) +
  scale_x_continuous(breaks = 1:5) +
  labs(title = "Distribución del Puntaje de Satisfacción",
       x = "Puntaje de Satisfacción",
       y = "Frecuencia") +
  theme_minimal()

# Imprimir resultados
for (i in 1:5) {
  cat(paste0("Proporción de clientes con satisfacción = ", i, ": ", round(propSatis[i], 4), "\n"))
  cat(paste0("Probabilidad de churn con satisfacción ", i, ": ", round(probChurnSatis[i], 4), "\n"))
  cat ("\n")
}
```

```r
promedioPointsChurn <- mean(df[df$Exited == 1, "Point.Earned"], na.rm = TRUE)
promedioPointsNoChurn <- mean(df[df$Exited == 0, "Point.Earned"], na.rm = TRUE)
cat(paste0("Promedio de Points Earned para churn: ", round(promedioPointsChurn, 4), "\n"))
cat(paste0("Promedio de Points Earned para no churn: ", round(promedioPointsNoChurn, 4), "\n"))
cat("\n")

rango <- seq(0, 1000, by = 100)
probChurnRango <- numeric((length(rango) - 1))

for (i in 1:(length(rango) - 1)) {
  clientes_rango <- df[df$Point.Earned >= rango[i] & df$Point.Earned < rango[i + 1], ]
  probChurnRango[i] <- sum(clientes_rango$Exited == 1) / nrow(clientes_rango)
}

# Imprimir resultados
for (i in 1:(length(rango) - 1)) {
  cat(paste0("Probabilidad de Churn para el rango (", rango[i], ", ", rango[i + 1], "): ", round(probChurnRango[i], 4), "\n"))
}
cat("\n")
# Data frame
resultados_rangos <- data.frame(
  Rango = paste0("[", rango[-length(rango)], ", ", rango[-1], ")"),
  ProbabilidadChurnRango = probChurnRango
)
resultados_rangos <- resultados_rangos[!is.na(resultados_rangos$ProbabilidadChurnRango), ]

#Histograma
ggplot(resultados_rangos, aes(x = Rango, y = ProbabilidadChurnRango)) +
  geom_bar(stat = "identity", fill = "steelblue", color = "black") +
  labs(
    title = "Probabilidad de Churn por Rango de Points Earned",
    x = "Rango de Points Earned",
    y = "Probabilidad de Churn"
  ) +
  theme_minimal()

clientesRango <- numeric(length(rango) - 1)

for (i in 1:(length(rango) - 1)) {
  clientesRango[i] <- sum(df$Point.Earned >= rango[i] & df$Point.Earned < rango[i + 1], na.rm = TRUE)
}

cat("Cantidad de clientes Points Earned:\n")
for (i in 1:(length(rango) - 1)) {
  cat(paste0("[", rango[i], ", ", rango[i + 1], "): ", clientesRango[i], "\n"))
}

cat("\n")
```

*¿Existe* una relación significativa entre la edad de los clientes y su probabilidad de abandonar el servicio?

```r
# Cargar la base de datos
data <- read.csv("Customer-Churn-Records.csv")


# Dividir los datos en dos grupos según Exited
age_no_churn <- data$Age[data$Exited == 0] # Clientes que no abandonaron
age_churn <- data$Age[data$Exited == 1]    # Clientes que abandonaron

# Realizar la prueba t
t_test_result <- t.test(age_churn, age_no_churn,
                        alternative = "two.sided",
                        conf.level = 0.99)

# Mostrar el resultado
print(t_test_result)

# Interpretación del resultado
if (t_test_result$p.value < 0.01) {
  print("Rechazamos la hipótesis nula: Hay una diferencia significativa en las edades.")
} else {
  print("No rechazamos la hipótesis nula: No hay evidencia de diferencia en las edades.")
}
```

Probabilidades de abandono

```r
# Calcular probabilidades condicionales
prob_churn_over_40 <- sum(data$Exited[data$Age > 40] == 1) / sum(data$Age > 40)
prob_churn_under_40 <- sum(data$Exited[data$Age <= 40] == 1) / sum(data$Age <= 40)

cat("Probabilidad de churn (edad > 40):", prob_churn_over_40, "\n")
cat("Probabilidad de churn (edad <= 40):", prob_churn_under_40, "\n")
```

La probabilidad de abandono para clientes con alto salario pero bajo credit score es del 22.30%, comparado con la tasa general de abandono del 20.38%.
```r
# Análisis de la variable "Card Type"
# Exploración inicial
cat("Tipos de tarjeta disponibles:", unique(df$Card.Type), "\n")
table_card_type <- table(df$Card.Type)
cat("Frecuencia de cada tipo de tarjeta:\n")
print(table_card_type)

# 1. Probabilidad Marginal
# Probabilidad de que un cliente tenga cada tipo de tarjeta
prob_card_type <- prop.table(table_card_type)
cat("Probabilidad de cada tipo de tarjeta:\n")
print(prob_card_type)

# 2. Probabilidad Condicional
# Probabilidad de churn dado un tipo de tarjeta
prob_churn_by_card <- df %>%
  group_by(Card.Type) %>%
  summarise(Prob_Churn = mean(Exited == 1))

cat("Probabilidad de churn por tipo de tarjeta:\n")
print(prob_churn_by_card)

# 3. Teorema de Bayes
# Probabilidad de que un cliente tenga una tarjeta específica dado que ha abandonado
p_exited <- mean(df$Exited == 1) # Probabilidad de churn
bayes_card_type <- df %>%
  group_by(Card.Type) %>%
  summarise(
    P_Exited_given_Card = mean(Exited == 1),
    P_Card = n() / nrow(df)
  ) %>%
  mutate(Bayes = (P_Exited_given_Card * p_exited) / P_Card)

cat("Probabilidad de cada tarjeta dado churn (Bayes):\n")
print(bayes_card_type)

# 4. Variables Aleatorias Discretas
# Esperanza matemática del tipo de tarjeta (asignando valores ordinales a las tarjetas)
df$Card.Type.Ordinal <- as.numeric(as.factor(df$Card.Type)) # Convertir a valores numéricos
expected_card_type <- mean(df$Card.Type.Ordinal)
cat("Esperanza matemática del tipo de tarjeta (ordinal):", expected_card_type, "\n")

# 5. Permutaciones y Combinaciones
# Número de formas de elegir 2 tipos de tarjeta de entre los disponibles
n_card_types <- length(unique(df$Card.Type))
r <- 2
combinations <- choose(n_card_types, r)
cat("Número de combinaciones de 2 tipos de tarjeta de entre", n_card_types, ":", combinations, "\n")

# 6. Prueba de Hipótesis (Chi-cuadrado)
# ¿Está el tipo de tarjeta asociado con la deserción?
chisq_test <- chisq.test(table(df$Card.Type, df$Exited))
cat("Prueba Chi-cuadrado de asociación entre Card Type y Exited:\n")
print(chisq_test)

# 7. Distribución Normal
# Analizando la puntuación promedio de satisfacción por tipo de tarjeta
satisfaction_by_card <- df %>%
  group_by(Card.Type) %>%
  summarise(
    Mean_Satisfaction = mean(Satisfaction.Score, na.rm = TRUE),
    SD_Satisfaction = sd(Satisfaction.Score, na.rm = TRUE)
  )
cat("Media y desviación estándar de la satisfacción por tipo de tarjeta:\n")
print(satisfaction_by_card)

# 8. Gráficos
# a. Probabilidad de churn por tipo de tarjeta
ggplot(prob_churn_by_card, aes(x = Card.Type, y = Prob_Churn, fill = Card.Type)) +
  geom_bar(stat = "identity") +
  labs(
    title = "Probabilidad de churn por tipo de tarjeta",
    x = "Tipo de Tarjeta",
    y = "Probabilidad de Churn"
  ) +
  theme_minimal()

# b. Satisfacción promedio por tipo de tarjeta
ggplot(satisfaction_by_card, aes(x = Card.Type, y = Mean_Satisfaction, fill = Card.Type)) +
  geom_bar(stat = "identity") +
  geom_errorbar(aes(ymin = Mean_Satisfaction - SD_Satisfaction,
                    ymax = Mean_Satisfaction + SD_Satisfaction), width = 0.2) +
  labs(
    title = "Satisfacción promedio por tipo de tarjeta",
    x = "Tipo de Tarjeta",
    y = "Satisfacción Promedio"
  ) +
  theme_minimal()

# 9. Variables Derivadas Relacionadas con "Card Type"
# a. Nivel de uso del tipo de tarjeta (frecuencia normalizada)
df <- df %>%
  group_by(Card.Type) %>%
  mutate(UsageLevel = n() / nrow(df)) %>%
  ungroup()

cat("Nivel de uso por tipo de tarjeta:\n")
print(unique(df[, c("Card.Type", "UsageLevel")]))

# b. Indicador de "Churn Risk" por tipo de tarjeta
df$ChurnRiskScore <- with(df, ifelse(Satisfaction.Score < 3 & Exited == 1, 1, 0))
churn_risk_by_card <- df %>%
  group_by(Card.Type) %>%
  summarise(AverageRisk = mean(ChurnRiskScore))

cat("Riesgo promedio de churn por tipo de tarjeta:\n")
print(churn_risk_by_card)

# c. Balance promedio por tipo de tarjeta
balance_by_card <- df %>%
  group_by(Card.Type) %>%
  summarise(Mean_Balance = mean(Balance, na.rm = TRUE))

cat("Balance promedio por tipo de tarjeta:\n")
print(balance_by_card)

# Gráfico de balance promedio por tipo de tarjeta
ggplot(balance_by_card, aes(x = Card.Type, y = Mean_Balance, fill = Card.Type)) +
  geom_bar(stat = "identity") +
  labs(
    title = "Balance promedio por tipo de tarjeta",
    x = "Tipo de Tarjeta",
    y = "Balance Promedio"
  ) +
  theme_minimal()
```

El análisis de la relación entre el tipo de tarjeta y el churn muestra que, aunque hay diferencias leves en la probabilidad de churn por tipo de tarjeta (DIAMOND tiene la probabilidad más alta con 21.8% y GOLD la más baja con 19.3%), el test de Chi-cuadrado indica que esta variable no tiene una asociación estadísticamente significativa con el churn (p > 0.05). Sin embargo, factores como el balance promedio más alto en clientes DIAMOND (~79,120) y el riesgo de churn ligeramente mayor en DIAMOND y SILVER podrían ser útiles para segmentaciones o estrategias combinadas con otras variables. En general, el tipo de tarjeta por sí solo no es un fuerte predictor del churn.

### **¿Cómo influye el número de productos adquiridos por un cliente en su probabilidad de churn, y qué implicaciones tiene esta relación en la estrategia del banco para diferentes mercados (Francia, España y Alemania)?**

```r
# Leer los datos
data <- read_csv("Customer-Churn-Records.csv")

# Asegurar que las variables sean factores donde corresponda
data$NumOfProducts <- as.factor(data$NumOfProducts)
data$Exited <- as.factor(data$Exited)

# --- ANÁLISIS ESTADÍSTICO Y PROBABILÍSTICO ---

# 1. Proporciones por Número de Productos
prop_products <- data %>%
  group_by(NumOfProducts) %>%
  summarise(
    Total = n(),
    Exited_Proportion = mean(as.numeric(as.character(Exited))),
    Not_Exited_Proportion = 1 - mean(as.numeric(as.character(Exited)))
  )

cat("Proporciones de clientes por Número de Productos:\n")
print(prop_products)

# 2. Probabilidad Total
# Probabilidad de churn considerando todos los números de productos
prob_total_churn <- data %>%
  summarise(Prob_Churn = mean(as.numeric(as.character(Exited))))

cat("Probabilidad total de churn:", prob_total_churn$Prob_Churn, "\n")

# 3. Probabilidad Condicional
# Probabilidad de churn dado un número específico de productos
prob_conditional <- data %>%
  group_by(NumOfProducts) %>%
  summarise(Prob_Churn = mean(as.numeric(as.character(Exited))))

cat("Probabilidad condicional de churn por Número de Productos:\n")
print(prob_conditional)

# 4. Teorema de Bayes
# Probabilidad de tener un número específico de productos dado que el cliente abandonó
p_exited <- mean(as.numeric(as.character(data$Exited)))
bayes_num_products <- data %>%
  group_by(NumOfProducts) %>%
  summarise(
    P_Exited_given_Products = mean(as.numeric(as.character(Exited))),
    P_Products = n() / nrow(data)
  ) %>%
  mutate(Bayes = (P_Exited_given_Products * p_exited) / P_Products)

cat("Probabilidad por Bayes de NumOfProducts dado churn:\n")
print(bayes_num_products)

# 5. Chi-cuadrado: Relación entre NumOfProducts y Churn
chisq_test <- chisq.test(table(data$NumOfProducts, data$Exited))
cat("Prueba Chi-cuadrado: Relación entre NumOfProducts y Churn:\n")
print(chisq_test)

# Interpretación de resultados
if (chisq_test$p.value < 0.05) {
  cat("Conclusión: Existe una relación significativa entre el Número de Productos y churn.\n")
} else {
  cat("Conclusión: No hay evidencia suficiente para afirmar que el Número de Productos y churn están relacionados.\n")
}

# --- VISUALIZACIONES ---

# Gráfico 1: Proporción de Clientes por Número de Productos y Estado de Retención
ggplot(data, aes(x = as.factor(NumOfProducts), fill = as.factor(Exited))) +
  geom_bar(position = "fill", width = 0.7) +
  scale_fill_manual(
    values = c("0" = "skyblue", "1" = "yellow"),
    labels = c("No Exited", "Exited")
  ) +
  labs(
    title = "Proporción de Clientes por Número de Productos y Estado de Retención",
    x = "Número de Productos",
    y = "Proporción de Clientes",
    fill = "Estado de Retención"
  ) +
  theme_minimal()


# Gráfico 3: Frecuencia de Productos por Estado de Retención
ggplot(data, aes(x = as.factor(NumOfProducts), fill = as.factor(Exited))) +
  geom_bar(position = "stack", width = 0.7) +
  labs(
    title = "Frecuencia de Productos por Estado de Retención",
    x = "Número de Productos",
    y = "Frecuencia",
    fill = "Estado de Retención"
  ) +
  theme_minimal()
```


1. **Distribución de clientes:**
   - La mayoría de los clientes tienen 1 o 2 productos (96.7%), con pocos clientes en las categorías de 3 (2.66%) y 4 productos (0.6%).
   - Esto sugiere que los clientes tienden a concentrarse en un portafolio limitado de productos bancarios.

2. **Probabilidad de churn por número de productos:**
   - Los clientes con 1 producto tienen una probabilidad de churn relativamente alta (27.7%).
   - Los clientes con 2 productos presentan la probabilidad de churn más baja (7.6%), lo que los convierte en el segmento más estable.
   - Los clientes con 3 productos tienen una probabilidad de churn alarmantemente alta (82.7%), mientras que todos los clientes con 4 productos abandonan (100% churn).
   - Este patrón podría indicar que clientes con múltiples productos enfrentan sobrecarga o insatisfacción.

3. **Relación significativa:**
   - La prueba Chi-cuadrado confirma que hay una asociación estadísticamente significativa entre el número de productos y el churn (p < 0.001). Este resultado es importante para el diseño de estrategias de retención basadas en el portafolio de productos.

4. **Análisis bayesiano:**
   - Dado el churn, los clientes con 3 y 4 productos tienen probabilidades muy altas (Bayes: 6.34 y 34.0 respectivamente) de pertenecer a estos segmentos. Esto refuerza la necesidad de monitorear de cerca a estos grupos.
